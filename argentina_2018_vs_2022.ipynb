```python
import pandas as pd
import numpy as np

df_wc = pd.read_csv("fifa_qatar.csv")

arg = df_wc[
    (df_wc["team1"] == "ARGENTINA") |
    (df_wc["team2"] == "ARGENTINA")
].copy()

arg["goles_a_favor"] = np.where(
    arg["team1"] == "ARGENTINA",
    arg["number of goals team1"],
    arg["number of goals team2"]
)

arg["goles_en_contra"] = np.where(
    arg["team1"] == "ARGENTINA",
    arg["number of goals team2"],
    arg["number of goals team1"]
)

arg[["team1", "team2", "goles_a_favor", "goles_en_contra"]]
```

```python
resumen_arg = arg[["goles_a_favor", "goles_en_contra"]].sum()
resumen_arg
```

Comparando la actuación de Argentina entre los Mundiales de 2018 y 2022, se observa una mejora muy marcada en el rendimiento ofensivo, ya que el equipo pasó de convertir 6 goles a 15. En defensa también hubo una leve mejora, reduciendo los goles recibidos de 9 a 8. Por lo tanto, la principal diferencia del equipo campeón en 2022 fue su mayor eficacia ofensiva, acompañada por un funcionamiento más equilibrado.

gráfico comparativo 2018 vs 2022

```python
import pandas as pd
import matplotlib.pyplot as plt

comparacion = pd.DataFrame({
    "Mundial": ["Rusia 2018", "Qatar 2022"],
    "Goles a favor": [6, 15],
    "Goles en contra": [9, 8]
})

comparacion.set_index("Mundial").plot(kind="bar", figsize=(8,5))
plt.title("Argentina: comparación 2018 vs 2022")
plt.ylabel("Cantidad de goles")
plt.grid()
plt.show()
```

Argentina mostró una mejora tanto ofensiva como defensiva entre 2018 y 2022. Sin embargo, la diferencia más importante se observa en el ataque, ya que aumentó notablemente la cantidad de goles convertidos, mientras que en defensa la mejora fue más leve.
